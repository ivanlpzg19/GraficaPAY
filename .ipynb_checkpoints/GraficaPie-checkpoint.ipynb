Crea una gráfica de barras que represente a los 10 países con más medallas ganadas. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv("C:\\Users\\ilopezg\\Downloads\\athlete_events.csv")
datos.head()
```

```python
data1=datos[['Year','NOC','Medal']]
data1.head()
```

```python
dummies=pd.get_dummies(data1[['Medal']])
dummies.head()
```

```python
concatena=pd.concat([data1,dummies], axis=1)
concatena.head()
```

```python
concatenar=concatena.groupby(['Year','Medal']).sum() #sumatoria a partir de dummies para las columnas que tienen valores numericos
concatenar.head()
```

```python
#concatenado=concatenar.groupby(['NOC','Medal']).sum() #sumatoria a partir de dummies para las columnas que tienen valores numericos
concatenar["total_medals"]=concatenar["Medal_Gold"]+concatenar["Medal_Silver"]+concatenar["Medal_Bronze"]# se crea la columna que contiene la sumatoria de las medalllas ganadas por País
concatenar.head()
```

```python
datasort=concatenar.sort_values("Year", ascending=False) #Se ordenan de mayor a menor los paises más ganadores
datasort.head()
```

```python
nombres=["Bronce","Oro","Plata"]
plot=datasort["total_medals"][:3].plot.pie(labels=nombres, autopct="%0.2f %%", figsize=(10,5))
plt.title("Gráfica de % de medallas ganadas en el último año")
plt.savefig('Gráfica de Pie.jpg')
```
